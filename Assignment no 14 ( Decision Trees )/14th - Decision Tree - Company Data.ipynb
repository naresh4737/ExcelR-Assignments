```python
x=data.iloc[:, 1:15]
x
```

```python
y=data['Sales']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
model = DecisionTreeClassifier(criterion = 'entropy', max_depth=3)
model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
fn=['CompPrice','Income', 'Advertising', 'Population', 'Price',  'Age', 'Education', 'ShelveLoc','Urban', 'US']
fig, axes = plt.subplots(nrows = 1, figsize = (4,4), dpi=300)
tree.plot_tree(model, feature_names = fn, filled = True);

plt.show()
```

```python
model_DecisionTreeClassifier = DecisionTreeClassifier(criterion = 'entropy', max_depth=3)  
model_DecisionTreeClassifier.fit(x_train, y_train)
```

```python
pred_test = model_DecisionTreeClassifier.predict(x_test)
pred_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred_test))
```
